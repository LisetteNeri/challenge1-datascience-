### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#1. Creamos listas 
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos_totales = []

# 2. Calcularmos la suma de la columna 'Precio' en cada DataFrame
for t in lista_tiendas:
    suma_precio = t['Precio'].sum()
    ingresos_totales.append(suma_precio)

# 3. Mostramos los resultados
print("Resumen de Ingresos Totales:")
for i in range(len(nombres_tiendas)):
    print(f"{nombres_tiendas[i]}: ${ingresos_totales[i]:,.2f}")
```

```python
# Configuramos el estilo y tamaño
plt.figure(figsize=(10, 6))
colores = ['#4CAF50', '#2196F3', '#FF9800', '#F44336'] # Un color para cada tienda

# Creamos el gráfico de barras
plt.bar(nombres_tiendas, ingresos_totales, color=colores)

# Añadimos detalles informativos
plt.title('Comparativa de Ingresos Totales por Tienda', fontsize=14)
plt.xlabel('Sucursales Alura Store', fontsize=12)
plt.ylabel('Ingresos (Suma de Precios)', fontsize=12)

# Mostramos el gráfico
plt.show()
```

# 2. Ventas por categoría

```python
# Obtenemos el conteo de ventas
def obtener_ventas_categoria(df):
    return df['Categoría del Producto'].value_counts()

# Diccionario para guardar los resultados
resumen_categorias = {}

print("--- Unidades vendidas por categoría ---")
for i, t in enumerate(lista_tiendas):
    nombre = nombres_tiendas[i]
    ventas = obtener_ventas_categoria(t)
    resumen_categorias[nombre] = ventas # Graficamos
    
    print(f"\nResultados de la {nombre}:")
    print(ventas)
```

```python
# Conteo global
df_total = pd.concat(lista_tiendas)
ventas_globales = df_total['Categoría del Producto'].value_counts()

# Gráfico Circular
plt.figure(figsize=(8, 8))
plt.pie(ventas_globales, 
        labels = ventas_globales.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

plt.title('Distribución Global de Ventas por Categoría', fontsize=15)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Promedio de calificaciones
def obtener_promedio_calificacion(df):
    return df['Calificación'].mean()

# Almacenar los promedios
medias_valoracion = []

print("--- Nivel de Satisfacción del Cliente ---")
for i, t in enumerate(lista_tiendas):
    promedio = obtener_promedio_calificacion(t)
    medias_valoracion.append(promedio)
    print(f"{nombres_tiendas[i]}: {promedio:.2f} estrellas")
```

```python
plt.figure(figsize=(10, 6))

colores_calidad = ['#27ae60' if x >= 4 else '#f1c40f' if x >= 3 else '#e74c3c' for x in medias_valoracion]
plt.barh(nombres_tiendas, medias_valoracion, color=colores_calidad)

plt.axvline(3.5, color='grey', linestyle='--', alpha=0.6, label='Meta de Calidad')

plt.title('Valoración Media por Sucursal', fontsize=14)
plt.xlabel('Estrellas (Promedio)', fontsize=12)
plt.xlim(0, 5) # La escala de calificación es de 1 a 5
plt.legend()
plt.grid(axis='x', alpha=0.3)

plt.show()
```

# 4. Productos más y menos vendidos

```python
def analizar_extremos_productos(df):
    conteo = df['Producto'].value_counts()
    mas_vendido = conteo.idxmax()
    cantidad_mas = conteo.max()
    menos_vendido = conteo.idxmin()
    cantidad_menos = conteo.min()
    return mas_vendido, cantidad_mas, menos_vendido, cantidad_menos

print(f"{'Tienda':<12} | {'Más Vendido':<20} (Cant) | {'Menos Vendido':<20} (Cant)")
print("-" * 80)

tops = []
lows = []

for i, t in enumerate(lista_tiendas):
    mv, cm, mnv, cmn = analizar_extremos_productos(t)
    tops.append(cm)
    lows.append(cmn)
    print(f"{nombres_tiendas[i]:<12} | {mv:<20} ({cm:2}) | {mnv:<20} ({cmn:2})")
```

```python

x = np.arange(len(nombres_tiendas))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, tops, width, label='Top Ventas', color='#2ecc71')
rects2 = ax.bar(x + width/2, lows, width, label='Mínimas Ventas', color='#e74c3c')

ax.set_ylabel('Unidades Vendidas')
ax.set_title('Brecha de Ventas: Producto Estrella vs. Producto Rezagado')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

plt.show()
```

# 5. Envío promedio por tienda

```python
# Promedio de una columna específica
def calcular_promedio_columna(df, nombre_columna):
    return df[nombre_columna].mean()

# Lista para almacenar los resultados
costos_envio_medios = []

print("--- Análisis de Costos de Logística (Pagados por el Cliente) ---")
for i, t in enumerate(lista_tiendas):
    promedio = calcular_promedio_columna(t, 'Costo de envío')
    costos_envio_medios.append(promedio)
    print(f"{nombres_tiendas[i]}: ${promedio:,.2f} promedio por pedido")
```

```python
plt.figure(figsize=(10, 6))

# Graficamos los puntos de cada tienda con un color distinto
colores = ['#3498db', '#e74c3c', '#2ecc71', '#f1c40f']

for i, t in enumerate(lista_tiendas):
    plt.scatter(t['Precio'], t['Costo de envío'], 
                alpha=0.5, 
                label=nombres_tiendas[i], 
                color=colores[i])

plt.title('Relación: Precio vs. Costo de Envío', fontsize=14)
plt.xlabel('Precio del Producto ($)', fontsize=12)
plt.ylabel('Costo de Envío ($)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
